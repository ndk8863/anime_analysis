```python
# anime_csv
anime_df = pd.read_csv(os.path.join(file_path,files['anime-dataset-2023']))
anime_df.head()
```

```python
# anime_df[anime_df['Score'] == 'UNKNOWN']
# unknownが存在することをどのように求めたのか→確認したらわかる
# score = unknownを除外
anime_df['Score'] = anime_df['Score'].replace('UNKNOWN','0.00')
anime_df['Score'] = anime_df['Score'].astype(float)

anime_df['Episodes'] = pd.to_numeric(anime_df['Episodes'].replace('UNKNOWN',-1),errors = 'coerce')
# errors='coerce　は、to_numericで変換できないときに、NaNとして扱われる

anime_df['Episodes'] = anime_df['Episodes'].fillna(-1).astype(int)

anime_df['Rank'] = pd.to_numeric(anime_df['Rank'],errors = 'coerce')
anime_df['Rank'] = anime_df['Rank'].fillna(-1).round().astype(int)

anime_df['Scored By'] = pd.to_numeric(anime_df['Scored By'],errors = 'coerce')
anime_df['Scored By'] = anime_df['Scored By'].fillna(-1).round().astype(int)

```
